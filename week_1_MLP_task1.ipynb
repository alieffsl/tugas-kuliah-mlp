```python
import pandas as pd

dataCSV = pd.read_csv('accelerometer.csv')
print(dataCSV)
```

```python
import numpy as np
import csv

with open('accelerometer.csv', 'r') as f:
    reader = csv.reader(f)
    data = list(reader)
    
data_array = np.array(data)
print(data_array)
```

```python
import matplotlib.pyplot as plt

pctid = 20

selected_data = dataCSV[dataCSV['pctid'] == pctid]

x = selected_data['x']
y = selected_data['y']
z = selected_data['z']

plt.plot(x, label='x', color='red')
plt.plot(y, label='y', color='green')
plt.plot(z, label='z', color='blue')

plt.legend()
plt.title('Grafik Vibrasi Accelerometer untuk PCTID = ' + str(pctid))
plt.show()
plt.figure(figsize=(30,10))
```
