Buatlah  sebuah  program  untuk  membuat  dataset  yang  terdiri  atas  5  variabel  dan  1 output kemudian plot data tersebut menggunakan 3 tipe visualisasi yang berbeda.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Membuat dataset
np.random.seed(42)
data = pd.DataFrame(np.random.rand(20, 6), columns=['var1', 'var2', 'var3', 'var4', 'var5', 'output'])

# Memplot data menggunakan scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data['var1'], data['output'], label='Var1')
plt.scatter(data['var2'], data['output'], label='Var2')
plt.scatter(data['var3'], data['output'], label='Var3')
plt.scatter(data['var4'], data['output'], label='Var4')
plt.scatter(data['var5'], data['output'], label='Var5')

plt.xlabel('Variabel')
plt.ylabel('Output')
plt.title('Scatter Plot Variabel vs Output')
plt.legend()
plt.show()

# Memplot data menggunakan line plot
plt.plot(data['var1'], data['output'], label='Var1')
plt.plot(data['var2'], data['output'], label='Var2')
plt.plot(data['var3'], data['output'], label='Var3')
plt.plot(data['var4'], data['output'], label='Var4')
plt.plot(data['var5'], data['output'], label='Var5')
plt.xlabel('Data')
plt.ylabel('Nilai')
plt.title('Line Plot')
plt.legend()
plt.show()

# Memplot data menggunakan heatmap
corr = data.corr()
plt.imshow(corr, cmap='coolwarm')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation=90)
plt.yticks(range(len(corr)), corr.columns)
plt.title('Heatmap')
plt.show()


# Memplot data menggunakan boxplot
plt.boxplot([data['var1'], data['var2'], data['var3'], data['var4'], data['var5']], labels=['Var1', 'Var2', 'Var3', 'Var4', 'Var5'])
plt.xlabel('Variabel')
plt.ylabel('Nilai')
plt.title('Boxplot')
plt.show()

```
