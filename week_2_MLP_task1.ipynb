```python
import pandas as pd
import matplotlib.pyplot as plt
import io
import requests

# Define the URL
url = "http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"

column_names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name']
c =pd.read_csv(url, sep='\s+', names=column_names,na_values=["?"])
```

```python
c.info()
```

```python
# Fungsi untuk mencetak statistik atribut dan menggambar boxplot
def print_and_plot_stats(attribute):
 mean = attribute.mean()
 maximum = attribute.max()
 minimum = attribute.min()
 mode = attribute.mode().iloc[0]
 missing = attribute.isnull().sum()
 print(f"Mean: {mean}")
 print(f"Max: {maximum}")
 print(f"Min: {minimum}")
 print(f"Mode: {mode}")
 print(f"Missing Value: {missing}\n")
 plt.figure(figsize=(8, 6))
 plt.boxplot(attribute, vert=False)
 plt.title(attribute.name)
 plt.show()
 
```

```python
# Loop untuk menganalisis setiap kolom atribut
for column in c.columns:
 if column != "car name":
     print(f"Statistik untuk Kolom '{column}':")
     print_and_plot_stats(c[column])
# Menghitung dan menampilkan jumlah missing value total dalam DataFrame
total_missing = c.isnull().sum().sum()
print(f"Total Missing Value di Seluruh DataFrame: {total_missing}")

```

```python
# Mengisi missing value pada kolom 'horsepower' dengan nilai modus
mode_horsepower = c['horsepower'].mode()[0]
c['horsepower'].fillna(mode_horsepower, inplace=True)
# Menampilkan jumlah missing value setelah pengisian
missing_value_horsepower_after_fill = c['horsepower'].isnull().sum()
print(f"Jumlah Missing Value pada 'horsepower' setelah pengisian: {missing_value_horsepower_after_fill}")
```
